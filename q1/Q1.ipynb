# Point 1

```
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
cov = [[1, 0.75], [0.75, 1]]
mu1 = [-3, 0]
mu2 = [3, 0]
mu3 = [0, 3]
N = 300
```

```
class myKMeans:
    
    def __init__(self, X, k=4):
        self.X = X
        self.K = k
        
    def fit(self):
        self.centroids = X[np.random.choice(len(X), self.K, replace=False)]
        self.intial_centroids = self.centroids
        self.prev_label,  self.labels = None, np.zeros(len(self.X))
        while not np.all(self.labels == self.prev_label) :
            self.prev_label = self.labels
            self.labels = self.predict(self.X)
            self.update_centroid(self.X)
        return self
        
    def predict(self, X):
        return np.apply_along_axis(self.compute_label, 1, X)

    def compute_label(self, x):
        return np.argmin(np.sqrt(np.sum((self.centroids - x)**2, axis=1)))

    def update_centroid(self, X):
        self.centroids = np.array([np.mean(X[self.labels == k], axis=0)  for k in range(self.K)])
```

# Point 2

```
# generates a gaussian random variable given mean and covariance 
def get_random_numbers(mu, cov, N):
  a = []
  for i in range(N):
    a.append(np.random.multivariate_normal(mu, cov))
  return a
```

```
def plot_cluster(cluster, X, y):
  plt.figure(figsize=(12,10))
  plt.title("Initial centers in black, final centers in red")
  plt.scatter(X[:, 0], X[:, 1], marker='.', c=y)
  plt.scatter(cluster.centroids[:, 0], cluster.centroids[:,1], c='r')
  plt.scatter(cluster.intial_centroids[:, 0], cluster.intial_centroids[:,1], c='k')
  plt.show()
```

```
cluster1 = get_random_numbers(mu1, cov, N)
cluster2 = get_random_numbers(mu2, cov, N)
cluster3 = get_random_numbers(mu3, cov, N)
```

```
# Plotting actual data
fig = plt.figure(figsize=(8,6))
X = []
X += [list(i) for i in cluster1]
X += [list(i) for i in cluster2]
X += [list(i) for i in cluster3]
X = np.array(X)
y = [1]*len(cluster1) + [2]*len(cluster2) + [3]*len(cluster3)
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Actual Data")
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.show()
```

## Actual and Predicted Values in clusters

### 2 Clusters

```
y_orig = [1]*len(cluster1) + [2]*len(cluster2) + [3]*len(cluster3)
cluster = myKMeans(X,2).fit()
plot_cluster(cluster, X, y_orig)
y = cluster.predict(X)
plot_cluster(cluster, X, y)
correct = 0
for i in range(len(y)):
  if y_orig[i] == y[i]:
    correct +=1
print("Accuracy for k=2 is:", correct/len(y_orig))
```
